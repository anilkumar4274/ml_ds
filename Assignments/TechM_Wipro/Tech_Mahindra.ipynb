# 2.Standard Deviation Method

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
df
```

```python
df.describe()
```

```python
plt.hist(df.No_of_trades, bins=20, rwidth=0.8)
plt.xlabel('No_of_trades')
plt.ylabel('VOLUME')
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
plt.hist(df.No_of_trades, bins=20, rwidth=0.8, density=True)
plt.xlabel('No_of_trades')
plt.ylabel('VOLUME')

rng = np.arange(df.No_of_trades.min(), df.No_of_trades.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df.No_of_trades.mean(),df.No_of_trades.std()))
plt.show()
```

```python
df.No_of_trades.mean()
```

```python
df.No_of_trades.std()
```

```python
upper_limit = df.No_of_trades.mean() + 3*df.No_of_trades.std()
upper_limit
```

```python
lower_limit = df.No_of_trades.mean() - 3*df.No_of_trades.std()
lower_limit
```

```python
df[(df.No_of_trades>upper_limit) | (df.No_of_trades<lower_limit)]
```

```python
df2 = df[(df.No_of_trades<upper_limit) & (df.No_of_trades>lower_limit)]
```

```python
df2
```

```python
df2.size
```

```python
df2.shape
```

```python
df2.describe()
```

# 4.Interquantile Method

```python
import pandas as pd
```

```python
df
```

```python
df.describe()
```

```python
q1 = df.No_of_trades.quantile(0.25)
q3 = df.No_of_trades.quantile(0.75)
q1, q3
```

```python
IQR = q3 - q1
IQR
```

```python
upper_lt = q3 + 1.5*IQR
lower_lt = q1 - 1.5*IQR
upper_lt, lower_lt
```

```python
df[(df.No_of_trades>upper_lt) | (df.No_of_trades<lower_lt)]
```

```python
df4 = df[(df.No_of_trades>lower_lt) & (df.No_of_trades<upper_lt)]
df4
```

```python
df4.shape
```

```python
df4.size
```

```python
df4.describe()
```
