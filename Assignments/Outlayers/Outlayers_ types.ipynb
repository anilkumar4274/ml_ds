# 2.Standard Deviation Method

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
df3 = pd.read_csv('Quote-Equity-TCS-EQ-21-10-2020-to-21-04-2021.csv')
df3
```

```python
df3.describe()
```

```python
plt.hist(df3.No_of_trades, bins=20, rwidth=0.8)
plt.xlabel('No_of_trades')
plt.ylabel('VOLUME')
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
plt.hist(df3.No_of_trades, bins=20, rwidth=0.8, density=True)
plt.xlabel('No_of_trades')
plt.ylabel('VOLUME')

rng = np.arange(df3.No_of_trades.min(), df3.No_of_trades.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df3.No_of_trades.mean(),df3.No_of_trades.std()))
plt.show()
```

```python
df3.No_of_trades.mean()
```

```python
df3.No_of_trades.std()
```

```python
upper_limit = df3.No_of_trades.mean() + 3*df3.No_of_trades.std()
```

```python
upper_limit
```

```python
lower_limit = df3.No_of_trades.mean() - 3*df3.No_of_trades.std()
```

```python
lower_limit
```

```python
df[(df3.No_of_trades>upper_limit) | (df3.No_of_trades<lower_limit)]
```

```python
df_no_outlayers = df[(df.No_of_trades>lower_limit) & (df.No_of_trades<upper_limit)]
df_no_outlayers
```

```python
df_no_outlayers.size
```

```python
df_no_outlayers.shape
```
