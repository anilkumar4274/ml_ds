# A heatmap is a plot of rectangular data as a color-encoded matrix. As parameter it takes a 2D dataset. That dataset can be coerced into an ndarray.

## This is a great way to visualize data, because it can show the relation between variabels including time.

#### HeatMap Using Seaborn

```python
import numpy as np
import seaborn as sns
np.random.seed(0)
sns.set_theme()
```

```python
# Uniform Distribution
uniform_data=np.random.rand(10,12)
ax=sns.heatmap(uniform_data,vmin=0,vmax=1)
```

```python
# Normal Distribution 
normal_data=np.random.randn(10,12)
ax=sns.heatmap(normal_data,center=0)
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
flights=flights.pivot("month","year","passengers")
```

```python
sns.heatmap(flights)
```

```python
ax=sns.heatmap(flights,annot=True,fmt='d',linewidths=.5)
```

```python
ax=sns.heatmap(flights,annot=True,fmt='d',cmap='YlGnBu')
```
